<p>1. create a copy of it</p>
<p>2. increment it </p>

<u><b>Vector operation</b></u>

<p>1) [] -> access value of element </p>
<p>2) = -> assign value to element at index i</p>
<p>3) push_back -> append item to the end of the vector</p>
<p>4) pop_back -> deletes the last item from the vector</p>
<p>5) insert -> inserts item at index i </p>
<p>6) erase -> erases item at index i </p>
<p>7) size -> returns the actual size used by elements </p>
<p>8) capacity -> returns the size of the allocated storage capacity</p>
<p>9) reserve -. requests a change in capacity to amount </p>

<p>In terms of sizing, there are two values of interest for a std::vector: size, and capacity (accessed via .size() and .capacity()).</p>
<p>.size() is the number of elements that are contained in the vector, whereas .capacity() is the number of elements that can be added to the vector, before memory will be re-allocated.</p>
<p>If you .push_back() an element, size will increase by one, up until you hit the capacity. Once the capacity is reached, most (all?) implementations, re-allocate memory, doubling the capacity.</p>

```c++
std::vector<int> A;
A.reserve(1);        // A: size:0, capacity:1  {[],x}
A.push_back(0);      // A: size:1, capacity:1  {[0]}
A.push_back(1);      // A: size:2, capacity:2  {[0,1]}
A.push_back(2);      // A: size:3, capacity:4  {[0,1,2],x}
A.push_back(3);      // A: size:4, capacity:4  {[0,1,2,3]}
A.push_back(4);      // A: size:5, capacity:8  {[0,1,2,3,4],x,x,x}
```

```c++
vector<int> intvector;
intvector.reserve(50); // reserves capacity of 50 at the start

for (int i = 0; i < 50; i++){
    intvector.push_back(i*i);
    std::cout << intvector[i] << std::endl;
}
```

Nevertheless, it is better to use the more precise term, hash table because other kinds of maps are sometimes implemented with a different underlying data structure.</p>
