### Mathematical operations

```c++
#include <math.h>
using namespace std;

 {
     cout << (2+3*4) << endl; //operations
     cout << (2+3)*4 << endl; //operations
     cout << pow(2, 3) << endl;
 }
```

```c++
#include <math.h>

std::cout << (2+3*4) << std::endl;
std::cout << (2+3)*4 << std::endl;
std::cout << pow(3, 3) << std::endl;
std::cout << 9/2 <<std::endl;
std::cout << 9%2 << std::endl;
std::cout << float(1)/3 << std::endl;
```

### Booleans

```c++
string strvar = "b"; //double quote marks for strings
char charvar = 'b'; //single quote marks for characters

std::cout << ("b" == strvar) << std::endl;
std::cout << ('b' == charvar) << std::endl;
```

### 2.3 Pointers

```c++
// c++ variable declarationa & assignment of an integer value
int varN =100;

std::cout << varN << std::endl;
std::cout << &varN << std::endl; //outputs the memory address of the variable
```

```c++
int varN1 = 9;
int *ptrn;
ptrn = &varN1
```

```c++
std::cout << ptrn <<std::endl;
std::cout << &varN1 << std::endl;
```

#### accessing values from pointers

```c++
std::cout << "varN1 value: " << varN1 << std::endl;
std::cout << "varN1 location: " << ptrn << std::endl;
std::cout << "dereference ptrn: " << *ptrn << std::endl;
```
