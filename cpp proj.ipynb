## Chap 3 - Control structures
### 3.1  Conditionals

```c++
//using else only

int grade = 85;

if (grade < 60) {
    std::cout << 'F' << std::endl;
}
else{
    if (grade < 70){
        std::cout << 'D' << std::endl;
    }
    else{
        if (grade <80){
            std::cout << 'C' << std::endl;
        }
        else{
            if (grade < 90){
                std::cout << 'B' << std::endl;
            }
            else{
                std::cout << 'A' << std::endl;
            }
        }
    }
}
```

```c++
//using else if

if (grade < 60){
    std::cout << 'F' << std::endl;
}

else if (grade < 70){
    std::cout << 'D' << std::endl;
}

else if (grade < 80){
    std::cout << 'C' << std::endl;
}

else if (grade < 90){
    std::cout << 'B' << std::endl;
}

else std::cout << 'A' << std::endl;
```

### 3.2 While Loops

```c++
int counter = 0;

while (counter <= 1){
    
    std::cout << "Hello world" << std::endl;
    counter = counter + 1;
}
```

## 4. Functions
### 4.1 defining c++ functions

```c++
int timesTwo(int num){
    return num * 2;
}
```

```c++
std::cout << timesTwo(5) << std::endl
```

```c++
double squareroot(double n){
    double root = n/2;
    for (int iii = 0; iii < 20; iii++){
        root = (0.5) * (root + (n / root));
    }
    
    return root;
}
```

```c++
std::cout << squareroot(9) << std::endl;
std::cout << squareroot(4563) << std::endl;
```

### Parameter passing: by Value v by Reference
