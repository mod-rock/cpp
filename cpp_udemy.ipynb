# Section 1
## 1.1 Array Basics

```c++
int A[10] = {2,4,6,8,40,50};

std::cout << sizeof(A) << std::endl; // each element is size 4, declared space is 10; so 10 x 4 = 40
```

```c++
std::cout << A[1] << std::endl;
std::cout << A[9] << std::endl; // the last element (i.e. 10) is not declared so it automatically becomes zero, while the second element is declared as 4
```

using for loop to print all elements in declared variable

```c++
int B[3] = {2,5,7};

for(int i = 0; i < 3; i++)
{
    std::cout << B[i] << std::endl;
}
```

```c++
int C[10] = {45,6,3,10};

for (int i = 0; i < 10; i++)
{
    std::cout << C[i] << std::endl;
}
```
