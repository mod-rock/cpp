## 1.3 Pointers
<p>The main memory of a computer is actually made up of 3 portions: 1) code section 2) the stack (which stores all data from the code section) 3) the heap (memory not accessible to the code section; the code section can only access memory from the stack).</p>
<p>The purpose of the pointer is to obtain the address of your desired memory (be it from the stack, the heap or any other sources of your data </p>
<p>hence you can indirectly retreive data using the pointer function</p>

```c++
#include <iostream>
```

```c++
int a = 1;
int *p; // initialise pointer

p =& a; // get the address from variable a (from stack memory)

std::cout << a << std::endl;
std::cout << *p << std::endl; // get address from a, then retrieve the value of a
std::cout << p << std::endl; // only get address from a
```

using pointers in recursion:

```c++
int A[5] = {2,4,6,8,10};
int *z;

z = A;

for (int i = 0; i < 5; i++)
{
    std::cout << A[i] << std::endl;
    std::cout << z[i] << std::endl;
}
```

storing data in heap memory:

```c++
#include <iostream> 

int *ph;

ph[0] = 10; ph[1] = 11; ph[2] = 12;

for (int i = 0; i < 3; i++)
{
    std::cout << ph[i] << std::endl;
}

delete [] ph;
```

## 1.4 references (i.e. another name for the variables s.t the same variable has 2 names)

```c++
#include <iostream>

int a = 10;
int &r = a;

std::cout << a << std::endl << r << std::endl;
```

r is the 'nickname' of a, but b is not the reference of a. In this case, value of r is changed to 30 and hence a's value is also changed to 30 since r is the nickname of a

```c++
int b = 30;

r = b;

std::cout << a << std::endl << r << std::endl;
```

## 1.5 pointer to structure

```c++
#include <iostream>

struct rectangle
{
    int length;
    int breadth;
};

rectangle r = {10,5}; // let r be rectangle

std::cout << r.length << std::endl << r.breadth << std::endl;
```

```c++
rectangle *p =& r; // let p be pointer of r

std::cout << p-> length << std::endl << p -> breadth << std::endl;
```

```c++
rectangle *z;

z  = new rectangle; //let compiler decide how much bytes to declare within heap

z -> length = 20;
z -> breadth = 15;

std::cout << z -> length << std::endl << z -> breadth << std::endl;
```

## 1.7 Parameter passing

```c++
#include <iostream>
```

```c++
int add_one (int a)
{
    a++;
    return a++;
}
```

```c++
int num3 = 10, num4 = 20;

std::cout << add_one(num3) << std::endl;
std::cout << num3 << std::endl;
```

```c++
void swap(int *a, int *b)
{
    int placeholder;
    placeholder = *b;
    *b = *a;
    *a = placeholder;
}

swap(&num3, &num4);

std::cout << num3 << std::endl << num4 << std::endl;
```
