## 1.3 Pointers
<p>The main memory of a computer is actually made up of 3 portions: 1) code section 2) the stack (which stores all data from the code section) 3) the heap (memory not accessible to the code section; the code section can only access memory from the stack).</p>
<p>The purpose of the pointer is to obtain the address of your desired memory (be it from the stack, the heap or any other sources of your data </p>
<p>hence you can indirectly retreive data using the pointer function</p>

```c++
#include <iostream>
```

```c++
int a = 1;
int *p; // initialise pointer

p =& a; // get the address from variable a (from stack memory)

std::cout << a << std::endl;
std::cout << *p << std::endl; // get address from a, then retrieve the value of a
std::cout << p << std::endl; // only get address from a
```

using pointers in recursion:

```c++
int A[5] = {2,4,6,8,10};
int *z;

z = A;

for (int i = 0; i < 5; i++)
{
    std::cout << A[i] << std::endl;
    std::cout << z[i] << std::endl;
}
```

storing data in heap memory:

```c++
#include <iostream> 

int *ph;

ph[0] = 10; ph[1] = 11; ph[2] = 12;

for (int i = 0; i < 3; i++)
{
    std::cout << ph[i] << std::endl;
}

delete [] ph;
```

## 1.9 Structure as parameter

```c++
//call by value, this method will not change the original values
#include <iostream>
```

```c++
struct distance
{
    int kilometer;
    int meter;
}       
```

```c++
void totaldistance(distance d1, distance d2)
{
    distance d;
    
    d.kilometer = d1.kilometer + d2.kilometer + (d1.meter + d2.meter) / 1000;
    d.meter = (d1.meter + d2.meter) % 1000;
    
    std::cout << "Total KM: " << d.kilometer << std::endl;
    std::cout << "total M: " << d.meter << std::endl;
}
```

```c++
distance Distance1, Distance2;

Distance1.kilometer = 10;
Distance1.meter = 455;

Distance2.kilometer = 9;
Distance2.meter = 745;

totaldistance(Distance1, Distance2);
```

<h2> Classes </h2>

```c++
#include <iostream>
```

```c++
class rectangle
{
    private:
        int length;
        int breadth;
    public:
        rectangle()
        {
            length = 0;
            breadth = 0;
        }
        
        rectangle(int l, int b)
        {
            length = l;
            breadth = b;
        }
        
        int area()
        {
            return length * breadth;
        }
        
        int perimeter()
        {
            return 2 * (length * breadth);
        }
        
        void setLength(int l)
        {
            length = l;
        }
        
        void setBreadth(int b)
        {
            breadth = b;}
    
        int getLength()
        {
            return length;}
    
        int getBreadth()
        {
            return breadth;}
        ~rectangle()
        {}
};
        
```

```c++
rectangle r(10,5);

std::cout << "area of rectangle is: " << r.area() << std::endl;
```

<h3>5.1 Static/global</h3>

```c++
#include <iostream>
#include <stdio.h>
```

```c++
int fun(int n)
{
    if(n > 0)
    {
        return fun(n - 1) + n;
    }
    return 0;
}
    
```

```c++
std::cout << fun(5);
```

```c++
int fun1(int n)
{
    if (n > 0)
    {
        static int x = 0; // static variable
        x++;
        return fun1(n - 1) + x;
    }
    return 0;
}
```

```c++
std::cout << fun1(5);
```

```c++
int x1 = 0;
int fun_2(int n)
{
    if (n > 0)
    {
        x1++;
        return fun_2(n - 1) + x1;
    }
    return 0;
}
```

<h3>5.2 Tail recursion</h3>

```c++
#include <iostream>
```

<b>Time</b> = order of n; <b>Space</b> = order of n:

```c++
void fun(int n)
{
    if (n > 0)
    {
        std::cout << n;
        fun(n - 1);
    }
}
```

<b>Time</b> = order of n; <b>Space</b> = order of 1 (because it doesn't call function at the end):

```c++
void loop_fun(int n)
{
    while (n > 0)
    {
        std::cout << n;
        n--;
    }
}
```

```c++
fun(4)
```

```c++
loop_fun(4)
```

<h3>5.4 Head recursion</h3>

```c++
#include <iostream>
```

```c++
void fun(int n)
{
    if (n > 0)
    {
        fun(n -1);
        std::cout << n;
    }
}
```

```c++
void loop_fun(int n)
{
    int i = 1;
    while (i <= n)
    {
        std::cout << i;
        i++;
    }
}
```

```c++
fun(4)
```

```c++
loop_fun(4)
```

<h3>5.5 Tree recursion</h3>

```c++
#include <iostream>
```

```c++
void tree_recursion(int n)
{
    if (n > 0)
    {
        std::cout << n;
        tree_recursion(n - 1);
        tree_recursion(n - 1);
    }
}
```

```c++
tree_recursion(2)
```

```c++
tree_recursion(3)
```

```c++
tree_recursion(4)
```

```c++
tree_recursion(5)
```
